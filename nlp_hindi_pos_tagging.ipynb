# Hindi POS Tagging (NLP)

This project demonstrates Part-of-Speech (POS) tagging for Hindi language text using Stanza (Stanford NLP).

Hindi POS tagging assigns grammatical roles such as noun, verb, adjective, and auxiliary to words in a sentence using trained language models.


```python
!pip install stanza
```

```python
import stanza
stanza.download('hi')
```

```python
nlp_hi = stanza.Pipeline(
    lang='hi',
    processors='tokenize,pos',
    use_gpu=False
)
```

```python
hindi_text = "प्राकृतिक भाषा प्रसंस्करण कंप्यूटर को मानव भाषा समझने में मदद करता है।"
```

```python
doc_hi = nlp_hi(hindi_text)

print("\nHindi POS Tagging:")
for sentence in doc_hi.sentences:
    for word in sentence.words:
        print(word.text, "->", word.upos)
```
